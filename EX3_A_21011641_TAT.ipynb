### Step 1. Import the necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/US_Baby_Names_right.csv).

### Step 3. Assign it to a variable called baby_names.

```python
baby_name = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/US_Baby_Names_right.csv')
```

```python
baby_name.head(4)
```

### Step 4. See the first 10 entries

```python
baby_name.head(10)
```

### Step 5. Delete the column 'Unnamed: 0' and 'Id'

```python
baby_names = baby_name.drop(['Unnamed: 0','Id'], axis = 1)
```

```python
baby_names
```

### Step 6. Is there more male or female names in the dataset?

```python
baby_names.Gender.value_counts()
```

```python
baby_names.Gender.describe()
```

### Step 7. Group the dataset by name and assign to names

```python
name = baby_names.groupby('Name')
```

```python
name
```

```python
name = baby_names.drop('Year', axis =1).groupby('Name').sum()
```

```python
name
```

### Step 8. How many different names exist in the dataset?

```python
baby_names.nunique()
```

### Step 9. What is the name with most occurrences?

```python
baby_names.groupby('Name').Count.sum().sort_values(ascending=False)
```

```python
name.Count.idxmax()
```

### Step 10. How many different names have the least occurrences?

```python
name[name.Count == name.Count.min()].shape[0]
```

### Step 11. What is the median name occurrence?

```python
(name.Count == name.Count.median()).sum()
```

### Step 12. What is the standard deviation of names?

```python
name.Count.std()
```
