### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv).

### Step 3. Assign it to a variable called chipo.

```python
data = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv', sep='\t')
```

```python
data.head(5)
```

### Step 4. How many products cost more than $10.00?

```python
#San pham hon $10.00
#Loai bo dau $ va chuyen doi thanh du lieu thich hop

data['item_price'] = data['item_price'].str.replace('$','')
data['item_price'] = data['item_price'].astype(float)
```

```python
np.count_nonzero(data.item_price > 10.00)
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
#Lay min gia mat hang chi cho 1 so luong gia
data_1 = data.groupby(by='item_name')[['item_price']].min().reset_index()
```

```python
data_1
```

### Step 6. Sort by the name of the item

```python
#sort theo ten mat hang
data.sort_values(by='item_name', ascending=True).reset_index(drop=True)
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
data_2 = np.where(data.item_price == data.item_price.max(), data.quantity, 0)
```

```python
for i in data_2:
  if i != 0:
    print("So luong mat hang Dat nhat dat la : " ,i)
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
data_3 = np.count_nonzero(data.item_name == 'Veggie Salad Bowl')
print(data_3,'lan Veggie Salad Bowl dat hang')
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
data_4 = np.count_nonzero((data.item_name == 'Canned Soda') & (data.quantity > 1))
```

```python
print(data_4,'lan dat hang nhieu hon 1 Canned Soda')
```
